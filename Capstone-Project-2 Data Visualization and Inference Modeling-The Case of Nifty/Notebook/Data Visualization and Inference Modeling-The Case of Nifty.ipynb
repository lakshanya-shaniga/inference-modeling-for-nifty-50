```python
# Given data of yearly returns
yearly_returns = [-14.65, -16.18, 3.25, 71.90, 10.68, 36.34, 39.83, 54.77, -51.79, 75.76, 17.95, -24.62, 27.70, 6.76, 31.39, -4.06, 3.01, 28.65, 3.15, 12.02, 14.90, 24.12, 4.32, 19.42]

# Calculate the mean and standard deviation
mean = np.mean(yearly_returns)
std_dev = np.std(yearly_returns)

# Define the buckets
buckets = {
    '< -2 std dev': [],
    '-2 to -1 std dev': [],
    '-1 to +1 std dev': [],
    '+1 to +2 std dev': [],
    '> +2 std dev': []
}

# Sort the yearly returns into the buckets
for return_value in yearly_returns:
    if return_value < mean - 2 * std_dev:
        buckets['< -2 std dev'].append(return_value)
    elif mean - 2 * std_dev <= return_value < mean - std_dev:
        buckets['-2 to -1 std dev'].append(return_value)
    elif mean - std_dev <= return_value <= mean + std_dev:
        buckets['-1 to +1 std dev'].append(return_value)
    elif mean + std_dev < return_value <= mean + 2 * std_dev:
        buckets['+1 to +2 std dev'].append(return_value)
    else:
        buckets['> +2 std dev'].append(return_value)

# Prepare data for plotting
labels = list(buckets.keys())
counts = [len(buckets[bucket]) for bucket in labels]

# Plot the histogram
plt.figure(figsize=(10, 6))
plt.bar(labels, counts, color='skyblue')
plt.xlabel('Standard Deviation Buckets')
plt.ylabel('Number of Yearly Returns')
plt.title('Histogram of Yearly Returns')
plt.show()
```

```python
# Select only the monthly return columns, excluding 'Year' and 'Annual'
monthly_returns = data.iloc[:, 1:-1].values.flatten()

# Calculate the mean and standard deviation of the monthly returns
mean = np.mean(monthly_returns)
std_dev = np.std(monthly_returns)

# Define the buckets based on standard deviation
buckets = {
    '< -2 std dev': [],
    '-2 to -1 std dev': [],
    '-1 to +1 std dev': [],
    '+1 to +2 std dev': [],
    '> +2 std dev': []
}

# Categorize each monthly return into the appropriate bucket
for return_value in monthly_returns:
    if return_value < mean - 2 * std_dev:
        buckets['< -2 std dev'].append(return_value)
    elif mean - 2 * std_dev <= return_value < mean - std_dev:
        buckets['-2 to -1 std dev'].append(return_value)
    elif mean - std_dev <= return_value <= mean + std_dev:
        buckets['-1 to +1 std dev'].append(return_value)
    elif mean + std_dev < return_value <= mean + 2 * std_dev:
        buckets['+1 to +2 std dev'].append(return_value)
    else:
        buckets['> +2 std dev'].append(return_value)

# Prepare data for plotting
labels = list(buckets.keys())
counts = [len(buckets[bucket]) for bucket in labels]

# Plot the histogram
plt.figure(figsize=(10, 6))
plt.bar(labels, counts, color='skyblue')
plt.xlabel('Standard Deviation Buckets')
plt.ylabel('Number of Monthly Returns')
plt.title('Histogram of Monthly Returns')
plt.show()
```
