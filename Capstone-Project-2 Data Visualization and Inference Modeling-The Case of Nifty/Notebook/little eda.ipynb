# Trend analysis
Trend analysis in the context of stock market data involves identifying and analyzing patterns or trends in the historical price movements of a particular security or index over time. 

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Fetch historical data for Nifty 50 index
nifty_data = yf.download('^NSEI', start='2020-01-01', end='2022-01-01')

# Plotting the closing price over time
plt.figure(figsize=(10, 6))
plt.plot(nifty_data['Close'], label='Nifty 50 Close Price', color='blue')
plt.title('Nifty 50 Close Price Trend')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.grid(True)
plt.show()

# Calculating and plotting moving averages
nifty_data['50MA'] = nifty_data['Close'].rolling(window=50).mean()
nifty_data['200MA'] = nifty_data['Close'].rolling(window=200).mean()

plt.figure(figsize=(10, 6))
plt.plot(nifty_data['Close'], label='Nifty 50 Close Price', color='blue')
plt.plot(nifty_data['50MA'], label='50-Day Moving Average', color='red')
plt.plot(nifty_data['200MA'], label='200-Day Moving Average', color='green')
plt.title('Nifty 50 Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

# Volatility analysis
Volatility analysis in the context of stock market data involves measuring the degree of variation of a trading price series over time. It's essential for assessing the risk associated with a particular security or index. 

```python
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt

# Fetch historical data for Nifty 50 index
nifty_data = yf.download('^NSEI', start='2020-01-01', end='2022-01-01')

# Calculate daily returns
nifty_data['Daily_Returns'] = nifty_data['Close'].pct_change()

# Calculate rolling standard deviation (volatility)
window_size = 20  # Adjust as needed
nifty_data['Volatility'] = nifty_data['Daily_Returns'].rolling(window=window_size).std() * np.sqrt(252)  # Annualized

# Plotting volatility
plt.figure(figsize=(10, 6))
plt.plot(nifty_data['Volatility'], label='Volatility', color='blue')
plt.title('Nifty 50 Volatility Analysis')
plt.xlabel('Date')
plt.ylabel('Volatility')
plt.legend()
plt.grid(True)
plt.show()
```

# Correlation analysis: 
Correlation analysis involves examining the relationship between two or more variables to determine how they move in relation to each other. In the context of stock market data, correlation analysis helps in understanding how the returns of different securities or indices are related. 

```python
import yfinance as yf
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Fetch historical data for Nifty 50 index
nifty_data = yf.download('^NSEI', start='2020-01-01', end='2022-01-01')

# Fetch historical data for another security (e.g., a stock)
# Replace 'AAPL' with the ticker symbol of the security you want to analyze
stock_data = yf.download('AAPL', start='2020-01-01', end='2022-01-01')

# Extract adjusted closing prices
nifty_close = nifty_data['Adj Close']
stock_close = stock_data['Adj Close']

# Combine into a single DataFrame
combined_data = pd.DataFrame({'Nifty 50': nifty_close, 'Stock': stock_close})

# Compute daily returns
returns = combined_data.pct_change().dropna()

# Calculate correlation matrix
correlation_matrix = returns.corr()

# Plot correlation matrix heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

# Outlier detection:
Outlier detection involves identifying data points that deviate significantly from the rest of the dataset. In the context of stock market data, outliers can represent abnormal price movements or errors in the data.

```python
import yfinance as yf
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Fetch historical data for Nifty 50 index
nifty_data = yf.download('^NSEI', start='2020-01-01', end='2022-01-01')

# Calculate daily returns
nifty_data['Daily_Returns'] = nifty_data['Close'].pct_change()

# Plot boxplot to visualize outliers
plt.figure(figsize=(8, 6))
sns.boxplot(x=nifty_data['Daily_Returns'])
plt.title('Boxplot of Nifty 50 Daily Returns')
plt.xlabel('Daily Returns')
plt.show()

# Calculate mean and standard deviation
mean_returns = nifty_data['Daily_Returns'].mean()
std_returns = nifty_data['Daily_Returns'].std()

# Set threshold for identifying outliers (e.g., 3 standard deviations from the mean)
threshold = 3 * std_returns

# Identify outliers
outliers = nifty_data[(nifty_data['Daily_Returns'] > mean_returns + threshold) | (nifty_data['Daily_Returns'] < mean_returns - threshold)]

print("Outliers:")
print(outliers)
```

# Seasonality analysis:
Seasonality analysis in stock market data involves identifying and analyzing recurring patterns or trends that occur at specific times of the year. These patterns may be influenced by factors such as seasonal changes in consumer behavior, economic cycles, or external events

```python
import yfinance as yf
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Fetch historical data for Nifty 50 index
nifty_data = yf.download('^NSEI', start='2010-01-01', end='2022-01-01')

# Convert index to datetime and extract month
nifty_data.index = pd.to_datetime(nifty_data.index)
nifty_data['Month'] = nifty_data.index.month

# Aggregate data by month and calculate mean monthly returns
monthly_returns = nifty_data.groupby('Month')['Close'].mean()

# Plot monthly returns to visualize seasonality
plt.figure(figsize=(10, 6))
sns.lineplot(x=monthly_returns.index, y=monthly_returns.values, marker='o')
plt.title('Mean Monthly Returns of Nifty 50')
plt.xlabel('Month')
plt.ylabel('Mean Monthly Returns')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.show()
```

# Statistical summaries:

Statistical summaries provide key insights into the distribution and characteristics of the data. In the context of stock market data analysis, statistical summaries can include measures such as mean, median, standard deviation, minimum, maximum, and percentiles# 

```python
import yfinance as yf

# Fetch historical data for Nifty 50 index
nifty_data = yf.download('^NSEI', start='2020-01-01', end='2022-01-01')

# Display basic statistical summary
print("Basic Statistical Summary:")
print(nifty_data['Close'].describe())

# Calculate additional statistics
median_price = nifty_data['Close'].median()
std_deviation = nifty_data['Close'].std()
min_price = nifty_data['Close'].min()
max_price = nifty_data['Close'].max()
percentile_25 = nifty_data['Close'].quantile(0.25)
percentile_75 = nifty_data['Close'].quantile(0.75)

# Display additional statistics
print("\nAdditional Statistics:")
print("Median Price:", median_price)
print("Standard Deviation:", std_deviation)
print("Minimum Price:", min_price)
print("Maximum Price:", max_price)
print("25th Percentile Price:", percentile_25)
print("75th Percentile Price:", percentile_75)
```
